### 2.2 Data 岗位数据

下面分析我最关心的数据相关岗位。首先，我通过标准职位和输入职位这两个字段，筛选我的目标岗位，即数据科学家，数据分析师和商业分析师

```python
# 筛选Data相关岗位
# 1) SOC_TITLE 精确匹配（大小写不敏感）
soc_targets = {s.casefold() for s in ["Data Scientists", "Business Intelligence Analysts"]}
mask_soc = main["SOC_TITLE"].astype(str).str.strip().str.casefold().isin(soc_targets)

# 2) JOB_TITLE 关键词子串匹配（大小写不敏感）
keywords = [
    "Data Analyst",
    "Data Scientist",
    "Data Science",
    "Business Analyst",
    "Data Analytics",
    "Advanced Analytics",
]
pattern = "|".join(re.escape(k) for k in keywords)
mask_job = main["JOB_TITLE"].astype(str).str.contains(pattern, case=False, na=False)

# 3) 合并条件并得到子集
data_subset = main[mask_soc | mask_job].copy()
```

2025财年，数据相关岗位的申请数量为33419（整体占比7%），总受益人50632人

```python
# Data相关岗位申请数量
total_cases = data_subset['CASE_NUMBER'].nunique()
print(f"Total number of cases submited: {total_cases}")

# Data相关岗位case对应的职位数量
print(f"Total positions applied: {data_subset['TOTAL_WORKER_POSITIONS'].astype(int).sum()}")
```

一般来说，面试流程可能1-3个月，2个月居多，那么可以看出，24年下半年数据类招聘较缓慢，直到25年1月才加速，高峰在25年3-4月

```python
# Data相关岗位申请数量，按月
plot_line_monthly_cases(data_subset)
```

Data岗位中，申请被批准的占比约98%；和整体相比，更少撤回申请的情况。

超过99%的Data相关申请都是H-1B类别，高于整体的不到98%，说明Data相关岗位更加依赖H-1B类别
